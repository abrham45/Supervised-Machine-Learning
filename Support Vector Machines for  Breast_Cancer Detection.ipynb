Scaling_Data

```python
Bcancer.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(Bcancer.drop('diagnosis',axis=1))
```

```python
B=scaler.transform(Bcancer.drop('diagnosis',axis=1))
```

```python
SBcancer=pd.DataFrame(B,columns=Bcancer.columns[1:])
```

```python
SBcancer.head()
```

Principal Component Analysis(PCA)

```python
from sklearn.decomposition import PCA
```

```python
P=PCA(n_components=10)
```

```python
P.fit(SBcancer)
```

```python
Bcancer_PCA=P.transform(SBcancer)
```

```python
Bcancer.shape
```

```python
Bcancer_PCA.shape
```

# Traning Our Data With Support Vector Machines

```python
X=Bcancer_PCA
y=Bcancer['diagnosis']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
mo=SVC()
```

```python
mo.fit(X_train,y_train)
```

```python
#prediction
prediction=mo.predict(X_test)
```

```python
prediction
```

# Evaluating the Model

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,prediction))
```

```python
cm=confusion_matrix(y_test,prediction)
```

```python
print(cm)
```

```python
sns.heatmap(confusion_matrix(y_test,prediction),annot=True)
```
